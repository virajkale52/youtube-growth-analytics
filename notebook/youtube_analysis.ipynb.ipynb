```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
f.head()
```

```python
X = d[["like_count"]]
Y = d[["view_count"]]
```

```python
#train test split
X_train , X_test , Y_train , Y_test  = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model_simple = LinearRegression().fit(X_train,Y_train)
model_simple
```

```python
#parameters
print("Slope (b1):", model_simple.coef_[0])
print("Intercept (b0):", model_simple.intercept_)
```

```python
#prdictions
y_pred = model_simple.predict(X_test)

mse  = mean_squared_error(Y_test, y_pred)
mae  = mean_absolute_error(Y_test, y_pred)
rmse = np.sqrt(mse)
r2   = r2_score(Y_test, y_pred)

print("MSE :", mse)
print("MAE :", mae)
print("RMSE:", rmse)
print("R¬≤  :", r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_log = d[["log_likes"]]
Y_log = d["log_views"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X_log, Y_log, test_size=0.2, random_state=42
)

model_log = LinearRegression()
model_log.fit(X_train, Y_train)

y_pred_log = model_log.predict(X_test)

r2_log = r2_score(Y_test, y_pred_log)
print("R¬≤ (Log Model):", r2_log)
```

| Model Type                 | R¬≤ Score |
| -------------------------- | -------- |
| Raw Linear Regression      | 0.55     |
| Log-Transformed Regression | 0.84     |


```python
X_log = d[["log_likes"]]
Y_log = d["log_views"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X_log, Y_log, test_size=0.2, random_state=42
)

model_log = LinearRegression()
model_log.fit(X_train, Y_train)

y_pred_log = model_log.predict(X_test)

r2_log = r2_score(Y_test, y_pred_log)
print("R¬≤ (Log Model):", r2_log)
```

The log-transformed model significantly improves predictive performance. This suggests that YouTube engagement follows exponential growth behavior rather than a simple linear pattern.

```python
import matplotlib.pyplot as plt

residuals_log = Y_test - y_pred_log

plt.scatter(y_pred_log, residuals_log)
plt.axhline(y=0, color='red')
plt.xlabel("Predicted Log Views")
plt.ylabel("Residuals")
plt.title("Residual Plot - Log Linear Regression")
plt.show()
```

The residual plot of the log-transformed model shows a more stable variance across predicted values. Compared to the raw model, heteroscedasticity is significantly reduced. This confirms that log transformation improved model assumptions.

A few extreme residuals suggest that viral videos may follow dynamics beyond engagement metrics alone.
